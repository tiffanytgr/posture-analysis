```python

import cv2
import mediapipe as mp

import pandas as pd
import seaborn as sns
import os
import csv
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

```python
# video_path = "/Users/anushaagarwal/Desktop/video1.mp4"
```

```python
import glob
import os.path

# good_dir = 'C:/Users/TTan29/OneDrive - Kenvue/Documents/Data Science/capstone-gait/data/good_pose'
# good_dir = 'D:/capstone-gait/data/good_pose/'
good_dir = 'E:/TT/GA_DSI/captone-running-form/data/good_pose/'
good_files = glob.glob(os.path.join(good_dir, '*.mp4')) + glob.glob(os.path.join(good_dir, '*.jpeg'))
print(good_files)

bad_dir = 'E:/TT/GA_DSI/captone-running-form/data/bad_pose/'
# bad_dir = 'D:/capstone-gait/data/bad_pose/'
bad_files = glob.glob(os.path.join(bad_dir, '*.mp4')) + glob.glob(os.path.join(bad_dir, '*.jpeg'))
print(bad_files)
```

```python
for video_path in good_files:
    
    print(video_path)

    cap = cv2.VideoCapture(video_path)
    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

        while cap.isOpened():
            ret, frame = cap.read()
            if ret:
                image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                image.flags.writeable = False

                results = holistic.process(image)
                image.flags.writeable = True
                image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

                mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                                         mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                         mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))

                cv2.imshow('Raw Webcam Feed', image)
            else:
                break

            if cv2.waitKey(10) & 0xFF == ord('q'):
                break
            
    print("exit")
    cap.release()
    cv2.destroyAllWindows()
    for i in range (1,5):
        cv2.waitKey(1)
    print("closed")
```

```python
num_coords = len(results.pose_landmarks.landmark)
landmarks = ['class']
for val in range(1, num_coords+1):
    landmarks += ['x{}'.format(val), 'y{}'.format(val), 'z{}'.format(val), 'v{}'.format(val)]
```

Good Poses

```python
with open('Pose_keypoints_good.csv', mode='w', newline='') as f:
    csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(landmarks)
```

```python
class_name = 'Good' #Left #Right #Down #Up #Nothing
```

```python
for video_path in good_files:
    
    print(video_path)
    #Video Feed

    # Setting up video capture device
    # 0 = built-in webcam & 2 = External webcam
    # If want to grab a video clip, type the name of video and its file loc within (...)

    cap = cv2.VideoCapture(video_path)

    # Start the holistic model
    # 'Holistic' allows me to access the Holistic estimation model
    # '0.6' = 60%

    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

        while cap.isOpened():
            ret, frame = cap.read()
#             print(ret)
            if ret == False:
                break
             #recoloring it back to BGR b/c it will rerender back to opencv
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False

            results = holistic.process(image)

            #setting image writeable back to true to be able process it
            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            #Draw the pose landmarks
            mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                                        mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                        mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))

            try:

                pose = results.pose_landmarks.landmark
                pose_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())

                row = pose_row
                row.insert(0, class_name)

                # data is appended to csv. if csv does not exist, new file is created
                with open('Pose_keypoints_good.csv', mode='a', newline='') as f:
                    csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                    csv_writer.writerow(row)

            except:
                pass
#                print("reaches here")
            cv2.imshow('Training Data Feed', image)
            
            


            if cv2.waitKey(10) & 0xFF == ord('q'):
                break
    print("exit")
    cap.release()
    cv2.destroyAllWindows()
    for i in range (1,5):
        cv2.waitKey(1)
    print("closed")
```

Visibility of Landmarks

```python
# Bar chart for visibility of landmarks
visibility_columns = df.filter(regex='^v', axis=1)
visibility_counts = visibility_columns.median()
plt.figure(figsize=(15, 6))
plt.bar(visibility_counts.index, visibility_counts.values)
plt.xlabel('Landmark Index')
plt.ylabel('Visibility %')
plt.title('Visibility of Landmarks')


plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D scatter plot for visibility, x1, and y1 coordinates
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['v1'], df['x1'], df['z1'])
ax.set_xlabel('Visibility')
ax.set_ylabel('x1 Coordinate')
ax.set_zlabel('y1 Coordinate')
ax.set_title('3D Scatter Plot: Visibility, x1, and y1')
plt.show()

```
